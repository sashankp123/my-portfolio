```python
#a)
city_name = "Amarillo"
print("City Name:", city_name)


# Assuming you have already set up the necessary API credentials and libraries
from googleapiclient.discovery import build

api_key = "AIzaSyBtNIc4G-ESzgoPPpVGHUF_0Xw0X0WG7nk"
youtube = build('youtube', 'v3', developerKey=api_key)

search_response = youtube.search().list(
    q=city_name,
    type="video",
    part="id,snippet",
#b)
    maxResults=30
).execute()

search_results = search_response.get("items", [])


for result in search_results:
    title = result['snippet']['title']
    video_id = result['id']['videoId']

    video_response = youtube.videos().list(
        part="statistics",
        id=video_id
    ).execute()

    views = video_response['items'][0]['statistics']['viewCount']
    likes = video_response['items'][0]['statistics']['likeCount']
    print("\nTitle:")
    print(title)
    print("Video Id:", video_id)
    print("\nNumber of Views:")
    print(views)
    print("Number of Likes:")
    print(likes)
    print("#############################")

print("\nCity Name:", city_name)
for result in search_results:
    title = result['snippet']['title']
    video_id = result['id']['videoId']

    video_response = youtube.videos().list(
        part="statistics",
        id=video_id
    ).execute()

    views = video_response['items'][0]['statistics']['viewCount']
    likes = video_response['items'][0]['statistics']['likeCount']
 #c), d)
    print("\nTitle:")
    print(title)
    print("Video Id:", video_id)
    print("\nNumber of Views:")
    print(views)
    print("Number of Likes:")
    print(likes)
    print("##########################")
    #graph code

# prompt: vizuvalize the data with the help of number of views and the number of likes
# e)I used bar method of graphical representation as it shows no.of views and likes in appropriate manner
#  and it shows views and likes in a same bar with different colors, In my graph in a bar, blue represents views and orange represents the viewers that liked it.

import matplotlib.pyplot as plt

views = []
likes = []
for result in search_results:
    video_id = result['id']['videoId']
    video_response = youtube.videos().list(part="statistics", id=video_id).execute()
    views.append(int(video_response['items'][0]['statistics']['viewCount']))
    likes.append(int(video_response['items'][0]['statistics']['likeCount']))

plt.bar(range(len(views)), views, label="Views")
plt.bar(range(len(likes)), likes, label="Likes")
plt.xlabel("Video")
plt.ylabel("Count")
plt.title("YouTube Video Views and Likes for {}".format(city_name))
plt.legend()
plt.show()
```
